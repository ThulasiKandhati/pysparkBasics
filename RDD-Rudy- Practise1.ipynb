RDD Operations Basics

```python
#map function
rdd = sc.parallelize(['b','c','a'])
sorted(rdd.map(lambda x:(x,1)).collect())
```

```python
#filter
rdd = sc.parallelize(['a','b','c'])
rdd.filter(lambda x:x == 'a').collect()
```

```python
#filter
rdd = sc.parallelize([1,10,34,45,9,13])
rdd.filter(lambda x:x%2 == 0).collect()
```

```python
rdd = sc.parallelize([1,3,4,10])
sorted(rdd.flatMap(lambda x: range(1,x)).collect())
```

```python
rdd.flatMap(lambda x: [(x,x),(x,x)]).collect()
```

```python
rdd = sc.parallelizeoka(("a",[1,2]),("b",[4,5]),("c",[9,8]))
def f(x): return x
x.flatMapValues(f).collect()
```

```python
m = sc.parallelize([(1,3),(4,5),(9,8)]).collectAsMap()
m[1]
```

Sampling/Filtering RDDs to pick out relevant Data points.

```python
from time import time
```

```python
sampled =  raw_data.sample(False,0.1,42)
contains_normal_sample = sampled.map(lambda line:line.split(",")).filter(lambda line:"normal" in line)
```

```python
t0 = time()
mum_sampled = contains_normal_sample.count()
duration =time()- t0
mum_sampled
```

Splitting DataSets and creating new combinations with set operations.

```python
normal_sample = sampled.filter(lambda line: "normal." in line)
```

```python
non_normal_sample = sampled.subtract(normal_sample)
```

```python
sampled.count()
```

```python
normal_sample.count()
```

```python
non_normal_sample.count()
```

```python
# cartesian finding common values between2 distinct features.
feature1 = sampled.map(lambda line:line.split(",")).map(lambda features:features[1]).distinct()
```

```python
feature2 = sampled.map(lambda line:line.split(",")).map(lambda features:features[2]).distinct()
feature0 = sampled.map(lambda line:line.split(",")).map(lambda features:features[0]).distinct()
```

```python
f1 = feature1.collect()
f2 = feature2.collect()
f0 = feature0.collect()
```

```python
len(feature1.cartesian(feature2).collect())
```
